```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
images=['images/Q2/1.jpg','images/Q2/2.jpg','images/Q2/3.jpg',
        'images/Q2/4.jpg','images/Q2/5.jpg','images/Q2/6.jpg',
        'images/Q2/7.jpg' , 'images/Q2/8.jpg']
imgs=[]
for i in range(len(images)):
    imgs.append(cv2.imread(images[i]))

for i in range(len(imgs)):
    k=i+1
    plt.subplot(2,4,k)
    a=imgs[i]
    plt.imshow(a)
    plt.axis('off')

    
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
st=cv2.Stitcher.create()
(dummy,outp)=st.stitch(imgs)
if dummy!=cv2.STITCHER_OK :
    print("error:(")
else:
    print("work successful")
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python
plt.imshow(outp)
plt.axis('off')
```
