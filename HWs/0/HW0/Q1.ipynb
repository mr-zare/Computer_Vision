```python
#import needed packages
import numpy as np
import matplotlib.image as mlimg
import matplotlib.pyplot as plt
import cv2
import os
```

```python
# read with cv2 and show image use cv2 and matplotlib Respectively 
image=cv2.imread("Q1.jpg")
##show with cv2 and matplotlib 
cv2.imshow("image",image)
plt.imshow(image)
plt.axis('off')
plt.show()
cv2.waitKey()
```

```python
##size image and channels
print(image.shape)
```

```python
#convert channel for equals image
conv=cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
#show with matplotlib
plt.imshow(conv)
plt.axis('off')
plt.show()
cv2.waitKey()
```

```python

#create directory if it doesn't exist
if not os.path.isdir('Q1/emptyfolder'):
    os.makedirs('Q1/emptyfolder')

image=[]
name=[]
shapes=[]
#read image from Q1 folder with grayscale 
#and save names and new images
for files in os.listdir('Q1'):
    if files.endswith('.jpeg'):
        print(files)
        r_image=cv2.imread(os.path.join('Q1/',files),0)
        image.append(r_image)
        fname=str(files).split('.')
        name.append(fname[0])
        


#save new image and write shapes in text file with same names as image's name
for i in range(len(image)):
    cv2.imwrite('Q1/emptyfolder/'+name[i]+'.jpeg' ,image[i])   
    with open('Q1/emptyfolder/'+name[i]+'.txt','w') as a :
        a.write(str(image[i].shape))
        a.close()
        
        

    
```
