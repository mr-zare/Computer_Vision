## Import dependencies

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Define functions

```python
def resize(image, size):
  '''
    calculate the resized image
    input(s):
      image (ndarray): input image
      size (tuple): size of output image
    output(s):
      images (ndarray): the resized image
  '''
  # resize image with opencv
  resized_image = cv2.resize(image,(size[0],size[1]))
  return resized_image

def crop(image, MainSize, CropSize):
  '''
    calculate the cropped images
    input(s):
      image (ndarray): input image
      MainSize (tuple): size of input image
      CropSize (tuple): size of output image
    output(s):
      images (ndarray): the cropped images
  '''
  # note that your output should be 1100 number of images
  # so you should not save the images that all their pixels have a value of zero
  # hint : you can sum all the pixels for each cropped image with numpy
  
  images = []
  #this is base index for x,y as heigh and width
  fx=0
  fy=0
  tx=MainSize[0]-CropSize[0] + 1
  ty=MainSize[1]-CropSize[1] + 1
  #now ccrop image with size of cropsize and add to images array
  #and check when all of 1110 small image exist return array
  # also update base index in every loop's cycle
  for i in range(tx):
    for j in range (ty):
      image_crop=image[fx:fx+CropSize[0],fy:fy+CropSize[1]]
      #for check that all of selected part isnt black 
      if np.sum(image_crop) !=0 :
        images.append(image_crop)
        if len(images) ==1100 :
          return images
    fy+=1
  fx+=1

  return images
```

## Read the images

```python
# read the images with opencv from Q2_images directory and add them to images array
images_path = ["usps_1.jpg","usps_2.jpg","usps_3.jpg","usps_4.jpg","usps_5.jpg"]
images = []
for i in range(5):
  #read pic from folder and add it to array
  image = cv2.imread(os.path.join('Q2_images/',images_path[i]),0)
  images.append(image) 
  assert image.shape == (544,528)
assert len(images) == 5
```

## Crop the images

```python
All_cropped_images = []
for img in images:
  MainSize = img.shape
  CropSize = (16,16)
  cropped_images = crop(img,MainSize,CropSize)
  All_cropped_images.append(cropped_images)

assert len(All_cropped_images) == 5
assert len(All_cropped_images[0]) == 1100
assert All_cropped_images[0][0].shape == (16,16)
```
